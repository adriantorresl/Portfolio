```python
import ee
import geemap
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import pandas as pd
import seaborn as sns
import requests
from io import BytesIO
import os
from PIL import Image  # Importación específica de PIL
```

```python
# Autenticación e inicialización de Google Earth Engine
try:
    ee.Initialize()
except Exception as e:
    ee.Authenticate()
    ee.Initialize()
```

```python
# 1. Definir el área de interés (Monterrey y zona metropolitana)
def get_monterrey_geometry():
    # Coordenadas aproximadas del área metropolitana de Monterrey
    monterrey = ee.Geometry.Polygon([
        [-100.45, 25.55],  # Esquina suroeste
        [-100.45, 25.85],  # Esquina noroeste
        [-99.95, 25.85],   # Esquina noreste
        [-99.95, 25.55]    # Esquina sureste
    ])
    return monterrey

aoi = get_monterrey_geometry()
```

```python
def get_cloud_free_composite(year):
    # Seleccionar colección según año
    if year <= 2013:
        collection_id = 'LANDSAT/LT05/C02/T1_L2'
        nir_band = 'SR_B4'
        red_band = 'SR_B3'
    elif 2013 < year <= 2021:
        collection_id = 'LANDSAT/LC08/C02/T1_L2'
        nir_band = 'SR_B5'
        red_band = 'SR_B4'
    else:
        collection_id = 'LANDSAT/LC09/C02/T1_L2'
        nir_band = 'SR_B5'
        red_band = 'SR_B4'
    
    # Filtrar colección
    collection = ee.ImageCollection(collection_id) \
        .filterDate(f'{year}-01-01', f'{year}-12-31') \
        .filterBounds(aoi) \
        .filter(ee.Filter.lt('CLOUD_COVER', 30))  # Filtrar imágenes con menos de 30% nubes
    
    # Función para procesar cada imagen
    def process_image(img):
        # Escalar valores
        optical = img.select(['SR_B.']).multiply(0.0000275).add(-0.2)
        # Máscara de nubes
        qa = img.select('QA_PIXEL')
        cloud_mask = qa.bitwiseAnd(1 << 3).Or(qa.bitwiseAnd(1 << 4)).Not()
        return optical.updateMask(cloud_mask).copyProperties(img, ['system:time_start'])
    
    # Procesar colección y crear mediana
    processed_collection = collection.map(process_image)
    
    # Verificar que hay imágenes disponibles
    count = processed_collection.size().getInfo()
    if count == 0:
        raise ValueError(f"No se encontraron imágenes para el año {year} con menos del 30% de nubes")
    
    return processed_collection.median().clip(aoi), nir_band, red_band
```

```python
# 3. Función mejorada para calcular NDVI
def calculate_ndvi(year):
    image, nir_band, red_band = get_cloud_free_composite(year)
    
    # Calcular NDVI usando nombres de bandas
    ndvi = image.normalizedDifference([nir_band, red_band]).rename('NDVI')
    return ndvi
```

```python
# 4. Obtener NDVI con manejo de errores
try:
    ndvi_2015 = calculate_ndvi(2015)
    print("NDVI 2015 generado correctamente")
except Exception as e:
    print(f"Error generando NDVI 2015: {str(e)}")
    ndvi_2015 = None

current_year = 2025  # Actualizar según necesidad
try:
    ndvi_current = calculate_ndvi(current_year)
    print(f"NDVI {current_year} generado correctamente")
except Exception as e:
    print(f"Error generando NDVI {current_year}: {str(e)}")
    ndvi_current = None
```

```python
# 5. Visualización (función actualizada)
def visualize_ndvi_comparison(ndvi_2015, ndvi_current, output_dir='ndvi'):
    
    os.makedirs(output_dir, exist_ok=True)
    # Configuración de visualización
    ndvi_params = {
        'min': -0.2,
        'max': 0.8,
        'palette': ['red', 'yellow', 'green']
    }
    
    # Crear mapa
    Map = geemap.Map(center=[25.6751, -100.3185], zoom=10)
    
    # Añadir capas al mapa (actualizado)
    Map.addLayer(ndvi_2015, ndvi_params, 'NDVI 2015')
    Map.addLayer(ndvi_current, ndvi_params, 'NDVI 2025')
    Map.addLayerControl()
    
    # Mostrar mapa
    display(Map)
    
    # Configuración común para las imágenes
    thumb_params = {
        'region': aoi,
        'dimensions': 1024,  # Aumenté la resolución
        'format': 'png',
        **ndvi_params
    }
    
    try:
        # 1. Guardar NDVI 2015
        url_2015 = ndvi_2015.getThumbUrl(thumb_params)
        response = requests.get(url_2015)
        with open(os.path.join(output_dir, 'ndvi_2015.png'), 'wb') as f:
            f.write(response.content)
        print(f"NDVI 2015 guardado en {os.path.join(output_dir, 'ndvi_2015.png')}")
        
        # 2. Guardar NDVI actual
        url_current = ndvi_current.getThumbUrl(thumb_params)
        response = requests.get(url_current)
        with open(os.path.join(output_dir, f'ndvi_{current_year}.png'), 'wb') as f:
            f.write(response.content)
        print(f"NDVI {current_year} guardado en {os.path.join(output_dir, f'ndvi_{current_year}.png')}")
        
        # 3. Crear y guardar comparación lado a lado
        plt.figure(figsize=(20, 10))
        
        # Cargar imágenes usando PIL
        img_2015 = Image.open(BytesIO(requests.get(url_2015).content))
        img_current = Image.open(BytesIO(requests.get(url_current).content))
        
        plt.subplot(1, 2, 1)
        plt.imshow(img_2015)
        plt.title('NDVI 2015', fontsize=14)
        plt.axis('off')
        
        plt.subplot(1, 2, 2)
        plt.imshow(img_current)
        plt.title(f'NDVI {current_year}', fontsize=14)
        plt.axis('off')
        
        comparison_path = os.path.join(output_dir, 'ndvi_comparison.png')
        plt.savefig(comparison_path, dpi=300, bbox_inches='tight')
        plt.close()
        print(f"Comparación guardada en {comparison_path}")
        
    except Exception as e:
        print(f"Error al generar o guardar imágenes: {str(e)}")

# Ejecutar visualización
visualize_ndvi_comparison(ndvi_2015, ndvi_current, output_dir='NDVI_results')
```

```python
# 6. Calcular diferencia y expansión urbana
def analyze_urban_expansion(ndvi_2015, ndvi_current, output_dir='urban_expansion_results'):
    
    # Crear directorio de salida si no existe
    os.makedirs(output_dir, exist_ok=True)
    # Calcular diferencia de NDVI
    ndvi_diff = ndvi_current.subtract(ndvi_2015).rename('NDVI_diff')
    
    # Identificar áreas con reducción significativa de NDVI (posible urbanización)
    urban_threshold = -0.2  # Umbral para considerar reducción significativa
    urban_expansion = ndvi_diff.lt(urban_threshold).rename('Urban_Expansion')
    
    # Configuración de visualización
    diff_params = {
        'min': -0.5,
        'max': 0.5,
        'palette': ['blue', 'white', 'red']
    }
    urban_params = {
        'palette': ['000000', 'FF0000'],  # Negro para 0, Rojo para 1
        'min': 0,
        'max': 1
    }
    
    
    # Calcular estadísticas
    stats = urban_expansion.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=aoi,
        scale=30,
        maxPixels=1e9
    )
    urban_percentage = stats.get('Urban_Expansion').getInfo() * 100
    print(f"Porcentaje de área con posible expansión urbana: {urban_percentage:.2f}%")
    
    # Crear mapa interactivo
    Map = geemap.Map(center=[25.6751, -100.3185], zoom=10)
    Map.addLayer(ndvi_diff, diff_params, 'Diferencia de NDVI')
    Map.addLayer(urban_expansion, urban_params, 'Expansión Urbana')
    Map.addLayerControl()
    display(Map)
    
    
    try:
        # Configuración común para thumbnails
        thumb_config = {
            'region': aoi,
            'dimensions': 1024,  # Mayor resolución
            'format': 'png'
        }
        
        # 1. Descargar y guardar imagen de diferencia
        diff_url = ndvi_diff.getThumbUrl({**thumb_config, **diff_params})
        response = requests.get(diff_url)
        diff_path = os.path.join(output_dir, 'diferencia_ndvi.png')
        with open(diff_path, 'wb') as f:
            f.write(response.content)
        print(f"Imagen de diferencia guardada en {diff_path}")
        
        # 2. Descargar y guardar imagen de expansión urbana
        urban_url = urban_expansion.getThumbUrl({**thumb_config, **urban_params})
        response = requests.get(urban_url)
        urban_path = os.path.join(output_dir, 'expansion_urbana.png')
        with open(urban_path, 'wb') as f:
            f.write(response.content)
        print(f"Imagen de expansión urbana guardada en {urban_path}")
        
        # 3. Crear y guardar imagen combinada
        plt.figure(figsize=(18, 8))
        
        # Cargar imágenes
        img_diff = Image.open(BytesIO(requests.get(diff_url).content))
        img_urban = Image.open(BytesIO(requests.get(urban_url).content))
        
        # Subplot 1: Diferencia de NDVI
        plt.subplot(1, 2, 1)
        plt.imshow(img_diff)
        plt.title('Diferencia de NDVI (Actual - 2015)', fontsize=12)
        plt.colorbar(label='Diferencia de NDVI', fraction=0.046, pad=0.04)
        plt.axis('off')
        
        # Subplot 2: Expansión urbana
        plt.subplot(1, 2, 2)
        plt.imshow(img_urban)
        plt.title(f'Áreas de Expansión Urbana ({urban_percentage:.2f}%)', fontsize=12)
        plt.colorbar(label='Expansión urbana (1=Sí, 0=No)', fraction=0.046, pad=0.04)
        plt.axis('off')
        
        combined_path = os.path.join(output_dir, 'analisis_completo.png')
        plt.savefig(combined_path, dpi=300, bbox_inches='tight')
        plt.close()
        print(f"Análisis combinado guardado en {combined_path}")
        
    except Exception as e:
        print(f"Error al descargar o guardar imágenes: {str(e)}")
    
    return urban_expansion

urban_expansion = analyze_urban_expansion(ndvi_2015, ndvi_current, output_dir='Expansion_Urbana_Monterrey')
```

```python
# Función para calcular estadísticas de NDVI
def calculate_ndvi_stats(ndvi_image, year, aoi):
    # Calcular histograma
    histogram = ndvi_image.reduceRegion(
        reducer=ee.Reducer.histogram(maxBuckets=20),
        geometry=aoi,
        scale=30,
        maxPixels=1e9
    ).getInfo()
    
    # Convertir a DataFrame
    hist_data = pd.DataFrame({
        'NDVI': histogram['NDVI']['bucketMeans'],
        'Frecuencia': histogram['NDVI']['histogram'],
        'Año': year
    })
    
    # Calcular estadísticas básicas
    stats = ndvi_image.reduceRegion(
        reducer=ee.Reducer.mean().combine(
            reducer2=ee.Reducer.median(),
            sharedInputs=True
        ).combine(
            reducer2=ee.Reducer.stdDev(),
            sharedInputs=True
        ),
        geometry=aoi,
        scale=30,
        maxPixels=1e9
    ).getInfo()
    
    stats['Año'] = year
    return hist_data, stats
```

```python
# Función para generar gráficos comparativos
def generate_comparison_charts(ndvi_2015, ndvi_current, aoi):
    # Calcular estadísticas para ambos años
    hist_2015, stats_2015 = calculate_ndvi_stats(ndvi_2015, 2015, aoi)
    hist_current, stats_current = calculate_ndvi_stats(ndvi_current, 2025, aoi)
    
    # Combinar datos para gráficos
    combined_hist = pd.concat([hist_2015, hist_current])
    combined_stats = pd.DataFrame([stats_2015, stats_current])
    
    # Configurar estilo de gráficos
    sns.set(style="whitegrid")
    plt.figure(figsize=(18, 12))
    
    # Gráfico 1: Histogramas comparativos
    plt.subplot(2, 2, 1)
    sns.barplot(data=combined_hist, x='NDVI', y='Frecuencia', hue='Año', palette="viridis")
    plt.title('Distribución de Valores NDVI')
    plt.xlabel('Valor NDVI')
    plt.ylabel('Frecuencia (píxeles)')
    plt.xticks(rotation=45)
    
    # Gráfico 2: Estadísticas comparativas
    plt.subplot(2, 2, 2)
    stats_melted = combined_stats.melt(id_vars=['Año'], var_name='Estadística', value_name='Valor')
    sns.barplot(data=stats_melted, x='Estadística', y='Valor', hue='Año', palette="magma")
    plt.title('Estadísticas de NDVI')
    plt.xlabel('Estadística')
    plt.ylabel('Valor')
    plt.xticks(rotation=45)
    
    # Gráfico 3: Diferencia de medias
    plt.subplot(2, 2, 3)
    diff = ndvi_current.subtract(ndvi_2015).rename('NDVI_diff')
    diff_stats = diff.reduceRegion(
        reducer=ee.Reducer.mean().combine(
            reducer2=ee.Reducer.median(),
            sharedInputs=True
        ).combine(
            reducer2=ee.Reducer.stdDev(),
            sharedInputs=True
        ),
        geometry=aoi,
        scale=30,
        maxPixels=1e9
    ).getInfo()
    
    diff_df = pd.DataFrame([diff_stats])
    diff_melted = diff_df.melt(var_name='Estadística', value_name='Valor')
    sns.barplot(data=diff_melted, x='Estadística', y='Valor', palette="coolwarm")
    plt.title('Diferencia de NDVI (2025 - 2015)')
    plt.xlabel('Estadística')
    plt.ylabel('Valor')
    plt.xticks(rotation=45)
    
    # Gráfico 4: Porcentaje de cambios
    plt.subplot(2, 2, 4)
    urban_threshold = -0.2
    urban_expansion = diff.lt(urban_threshold)
    stats = urban_expansion.reduceRegion(
        reducer=ee.Reducer.mean(),
        geometry=aoi,
        scale=30,
        maxPixels=1e9
    ).getInfo()
    
    urban_percentage = stats['NDVI_diff'] * 100
    other_percentage = 100 - urban_percentage
    
    data = pd.DataFrame({
        'Categoría': ['Expansión Urbana', 'Otras Áreas'],
        'Porcentaje': [urban_percentage, other_percentage]
    })
    
    sns.barplot(data=data, x='Categoría', y='Porcentaje', palette="Set2")
    plt.title('Porcentaje de Expansión Urbana')
    plt.xlabel('Categoría')
    plt.ylabel('Porcentaje (%)')
    
    plt.tight_layout()
    plt.savefig('ndvi_statistical_comparison.png', dpi=300, bbox_inches='tight')
    plt.show()
```

```python
# Llamar a la función después de tus análisis existentes
if ndvi_2015 is not None and ndvi_current is not None:
    generate_comparison_charts(ndvi_2015, ndvi_current, aoi)
else:
    print("No se pueden generar gráficos comparativos porque falta una o ambas imágenes NDVI")
```
